Consider a vector X containing input features and Y as teh target Assume hat the input feature X is multi-dimensional.Fit a linear regression model using the given Input X and the target Y
After fitting, obtain the intercept of the fitted linear regrssion model as output

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
import math
```

```python
X = [[1,2],[3,4],[5,1],[1,12],[12,11]]
Y= [6,8,9,10,1]
```

```python
X_data = np.array(X).reshape(-1,1)
```

```python
linreg = LinearRegression()
linreg.fit(X,Y)
```

```python
Y_data = np.array(Y).reshape(-1,1)
```

```python
predict = linreg.predict(X)
```

```python
predict
```

```python
from matplotlib import pyplot as plt
plt.scatter(Y,predict, color="green")
plt.plot([0,15],[0,15], color="red")
plt.title("Linear Regression")
```

```python
print(linreg.intercept_)
```

```python
print(f"{linreg.score(X,Y)}")
```

```python
intercept = linreg.predict([[0,0]])
print("Intercept: ",intercept[0])
```
