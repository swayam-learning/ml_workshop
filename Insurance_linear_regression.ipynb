```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
insurance = pd.read_csv("insurance.csv")
insurance
```

```python
age_data = insurance["age"]
charges_data = insurance["charges"]
```

```python
age = np.array(age_data).reshape(-1,1)
charges = np.array(charges_data).reshape(-1,1)
```

```python
linreg = LinearRegression()
linreg.fit(age,charges)
```

```python
prediction = linreg.predict(age)
```

```python
from matplotlib import pyplot as plt
plt.scatter(age,charges, color="green")
plt.plot(age,prediction, color="red")
plt.title("Linear Regression")
plt.xlabel("Age")
plt.ylabel("Charges")
```
