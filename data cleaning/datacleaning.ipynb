```python
import pandas as pd
pd.set_option("display.max_column",None)
```

```python
df = pd.read_csv("Uncleaned_DS_jobs.csv")
df
```

```python
df.info()
```

```python
df["Job Description"]=df["Job Description"].str.replace("\n","")
```

```python
df["Job Description"]
```

```python
df.head()
```

```python
df["Salary Estimate"]=df["Salary Estimate"].str.replace("$","")
```

```python
df["Salary Estimate"]=df["Salary Estimate"].str.replace("K","")
```

```python
df.head()
```

```python
df["Salary Estimate"]=df["Salary Estimate"].str.replace("(Glassdoor est.)","")
```

```python
df.head()
```

```python
df.drop("index",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["min_salary"] = df["Salary Estimate"][0][0:3]
```

```python
df["max_salary"] = df["Salary Estimate"][0][4:7]
```

```python
df.head()
```

```python
df["min_salary"] = df["min_salary"].astype(int)
```

```python
df["max_salary"] = df["max_salary"].astype(int)
```

```python
df["avg_salary"] = (df["min_salary"]+df["max_salary"])/2
```

```python
df.head()
```

```python
df["Company Name"]=df["Company Name"].str.split('\n').str[0]
```

```python
df.head()
```

```python
df.drop("Competitors",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["avg_salary"]=df["avg_salary"].astype(int)
```

```python
df.head()
```

```python
df["Job State"] = df["Location"].str.split(",").str[-1]
```

```python
df.head()
```

```python
df["Same State"] = (df["Location"].str.split(",").str[0]) == (df["Headquarters"].str.split(",").str[0])
```

```python
df["Same State"]=df["Same State"].astype(int)
```

```python
df.head()
```

```python
df["Company Age"] = 2024- df["Founded"].astype(int)
```

```python
df.head()
```

```python
import re
languages = ['Python','Excel','Hadoop', 'Spark', 'AWS','Tableau', 'Big_Data']
pattern = r'\b(?:' + '|'.join(languages) + r')\b'
df['programming_languages'] = df['Job Description'].apply(lambda x: ', '.join(re.findall(pattern, x, flags=re.IGNORECASE)))
```

```python
df.head()
```

```python
df.drop_duplicates(subset="programming_languages")
```

```python
# extract name of technologies from job description column
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['hadoop'] = df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['big_data'] = df['Job Description'].apply(lambda x: 1 if 'big_data' in x.lower() else 0)
```

```python
df.head()
```

```python
df.groupby("Job Title")["Job Title"].value_counts()
```

```python
df["Job Title"].values
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
df['job_simp'] = df['Job Title'].apply(title_simplifier)
df.job_simp.value_counts()
```
