```python
import pandas as pd
```

```python
df = pd.read_csv("Uncleaned_DS_jobs.csv")
df
```

```python
df.info()
```

```python
df["Job Description"]=df["Job Description"].str.replace("\n","")
```

```python
df["Job Description"]
```

```python
df.head()
```

```python
df["Salary Estimate"]=df["Salary Estimate"].str.replace("$","")
```

```python
df["Salary Estimate"]=df["Salary Estimate"].str.replace("K","")
```

```python
df.head()
```

```python
df["Salary Estimate"]=df["Salary Estimate"].str.replace("(Glassdoor est.)","")
```

```python
df.head()
```

```python
df.drop("index",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["min_salary"] = df["Salary Estimate"][0][0:3]
```

```python
df["max_salary"] = df["Salary Estimate"][0][4:7]
```

```python
df.head()
```

```python
df["min_salary"] = df["min_salary"].astype(int)
```

```python
df["max_salary"] = df["max_salary"].astype(int)
```

```python
df["avg_salary"] = (df["min_salary"]+df["max_salary"])/2
```

```python
df["Company Name"]=df["Company Name"].str.split('\n').str[0]
```
