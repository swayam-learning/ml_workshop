```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np
```

```python
dataset = np.array([["Salty","Hot","Soft","No"],
                   ["Spicy","Hot","Soft","No"],
                   ["Spciyy","Hot","Hard","Yes"],
                   ["Spicy","Cold","Hard","No"],
                   ["Spicy","Hot","Hard","Yes"],
                   ["Sweet","Cold","Soft","Yes"],
                   ["Salty","Cold","Soft","No"],
                   ["Sweet","Hot","Soft","Yes"],
                   ["Spicy","Cold","Soft","Yes"],
                   ["Salty","Hot","Hard","Yes"]])
```

```python
features = dataset[:,:-1]
targets = dataset[:,-1]
```

```python
enc = preprocessing.OrdinalEncoder()
encoded_features = enc.fit_transform(features)
```

```python
encoded_features
```

```python
feature_train,feature_test,target_train,target_test = train_test_split(encoded_features,targets,test_size=0.10)
```

```python
guess = np.array([["Salty","Hot","Hard"]])
```

```python
encoded_guess = enc.transform(guess)
```

```python
encoded_guess
```

```python
model = DecisionTreeClassifier()
fitted = model.fit(feature_train,target_train)
```

```python
predictions = fitted.predict(encoded_guess)
```
